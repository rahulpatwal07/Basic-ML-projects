```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CarPrice_Assignment.csv")
```

```python
df.head()
```

```python
df.drop("car_ID",axis=1,inplace=True)
```

```python
df
```

```python
df["CompanyName"] = df.CarName.apply(lambda x: x.split()[0])
```

```python
df
```

```python
df.drop(columns="CarName", inplace=True)
```

```python
df.head()
```

```python
df['CompanyName'].unique().tolist()
```

```python
df["CompanyName"].replace("maxda","mazda", inplace=True)
df["CompanyName"].replace("Nissan","nissan", inplace=True)
df["CompanyName"].replace("porcshce","porsche", inplace=True)
df["CompanyName"].replace("toyouta","toyota", inplace=True)
df["CompanyName"].replace("porcshce","porsche", inplace=True)
df["CompanyName"].replace("vokswagen","volkswagen", inplace=True)
df["CompanyName"].replace("vw","volkswagen", inplace=True)
```

```python
df['CompanyName'].unique().tolist()
```

```python
df
```

```python
categorical_features = df.dtypes == object
categorical_columns = df.columns[categorical_features].tolist()
```

```python
categorical_columns
```

```python
pd.get_dummies(df)
```

```python
X = pd.get_dummies(df, dtype=int)
```

```python
X.head()
```

```python
y = X['price']
```

```python
X.drop('price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2Score = r2_score(y_test,y_pred)
print("RMSE : ",rmse)
print("R2 score : ",r2Score)
```

```python
plt.figure(figsize=(10,  6),dpi=200)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import Ridge
model2 = Ridge()
model2.fit(X_train,y_train)
y_pred2 = model2.predict(X_test)
rmse2 = np.sqrt(mean_squared_error(y_test, y_pred2))
r2Score2 = r2_score(y_test,y_pred2)
print("RMSE : ",rmse2)
print("R2 score : ",r2Score2)
```

```python
plt.figure(figsize=(10,  6),dpi=200)
plt.scatter(y_test, y_pred2, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import Lasso
model3 = Lasso()
model3.fit(X_train,y_train)
y_pred3 = model3.predict(X_test)
rmse3 = np.sqrt(mean_squared_error(y_test, y_pred3))
r2Score3 = r2_score(y_test,y_pred3)
print("RMSE : ",rmse3)
print("R2 score : ",r2Score3)
```

```python
plt.figure(figsize=(10,  6),dpi=200)
plt.scatter(y_test, y_pred3, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.grid(True)
plt.show()
```
